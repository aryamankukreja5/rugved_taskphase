```python
#Ques18
import csv
x = []
y = []
  
with open('rugved_task4/matches.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
      
    for row in plots:
        x.append(row[7])
        y.append(row[7])
  
plt.bar(x, y, color = 'b', width = 0.72, label ="toss decision")
plt.xlabel('Decision')
plt.ylabel('Number')
plt.title('Toss Decision')
#plt.legend()
plt.show()
```
